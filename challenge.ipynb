```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01',
        '2020-02-01',
        '2021-03-01',
        '2022-04-01',
        '2023-05-01',
        '2023-06-01',
        '2020-07-01',
        '2021-08-01',
        '2022-09-01',
        '2023-10-01',
        '2022-11-01',
        '2023-12-01',
        ],
    'valor': [
        400000,
        890000,
        760000,
        430000,
        920000,
        340000,
        800000,
        500000,
        200000,
        900000,
        570000,
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)

# convertendo data_ref para melhor visualização:
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])

# Arranja por 'data_ref'
df_faturamento = df_faturamento.sort_values(by='data_ref')

# Média
media_vendas = round(df_faturamento['valor'].mean(), 2)

media_vendas
```

```python
bars = plt.bar(df_faturamento['data_ref'], df_faturamento['valor'], color='red', width = 40)

# Adiciona título e labels
plt.title('Faturamento Mensal', fontsize=16)
plt.xlabel('Data de Referência', fontsize=14)
plt.ylabel('Valor (R$)', fontsize=14)

# Rotaciona o eixo-x para ler melhor
plt.xticks(rotation=45)

# Adiciona grid
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
bars = plt.plot(df_faturamento['data_ref'], df_faturamento['valor'], color='red', marker='o', linestyle='-', linewidth=2)

# Adiciona título e labels
plt.title('Faturamento Mensal', fontsize=16)
plt.xlabel('Data de Referência', fontsize=14)
plt.ylabel('Valor (R$)', fontsize=14)

# Rotacona o eixo-x para ler melhor
plt.xticks(rotation=45)

# Adiciona grid
plt.grid(axis='y', linestyle='--', alpha=0.7)
```
